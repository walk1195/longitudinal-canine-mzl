```python
# Quick run through standard processing
adata = adata_sc
adata.layers["counts"] = adata.X.copy()
print('Normalizing data...')
sc.pp.normalize_total(adata, inplace=True)
sc.pp.log1p(adata)
adata.layers["log_norm"] = adata.X.copy()
print('Running PCA...')
sc.pp.pca(adata)
print('Generating neighbors graph...')
sc.pp.neighbors(adata, n_pcs=30) # Neighbors graph
print('Calculating umap embeddings...')
sc.tl.umap(adata)
print('Plotting cell types...')
sc.pl.umap(adata, color='cell_type_final') # Test plot
```
