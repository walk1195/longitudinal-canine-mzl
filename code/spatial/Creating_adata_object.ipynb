## Converting seurat object to adata

This script reads in matrix, cell, and feature .csvs from a deconstructed Seurat single-cell object for input into adata

Single-cell adata object is saved for future use to deconvolute spatial

```python
# Import libs
from pathlib import Path
import os

import scanpy as sc
import squidpy as sq
import anndata as ad
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sc.logging.print_header()
print(f"squidpy=={sq.__version__}")
```

```python
import scipy.io as io
```

```python
# Set wkdir
data_dir = ''
os.chdir(data_dir)
```

```python
# Read in counts, obs, and feature files
counts = io.mmread('orv01_counts_matrix.mtx')
cell_meta = pd.read_csv('orv01_obs.csv', index_col=0)
gene_meta = pd.read_csv('orv01_genes.csv', index_col=0)
```

```python
cell_meta
```

```python
gene_meta
```

```python
# Create adata object
adata = ad.AnnData(X=counts.transpose().tocsr()) # Need to transpose matrix and convert to CSR format
adata.obs = cell_meta
adata.var = gene_meta
```

```python
adata.var
```

```python
# Fix adata.var
# Move gene names from col to index
adata.var.index = adata.var["x"].astype(str)  # Ensure it's a string
adata.var.drop(columns=["x"], inplace=True)  # Remove the old column
```

```python
# Save as single-cell adata object for later use
adata.write(f'{data_dir}adata_object_orv01.h5ad')
```
